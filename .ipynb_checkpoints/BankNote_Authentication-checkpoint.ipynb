```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('BankNote_Authentication.csv')
data.head()
```

```python
data.isnull().any()
```

```python
data.describe()
```

```python
sns.countplot(data = data, x = 'class')
```

**There is enough data in both classes to make classifications properly. There is no biase.**

```python
fig, ax = plt.subplots(2, 2)
plt.figure(figsize=(10,6))
fig.tight_layout(pad=2.0)
k = 0
for i in range(0, 2):
    for j in range(0, 2):
        sns.histplot(data = data, x = data.columns[k], kde = True, ax = ax[i,j])
        k+=1
```

```python
sns.heatmap(data.corr(), annot = True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.drop("class", axis = 1), data['class'], test_size = 0.3, random_state = 42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
randomForest = RandomForestClassifier()
randomForest.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = randomForest.predict(X_test)
score = accuracy_score(y_test, y_pred)
score
```

```python
from sklearn import metrics

print(metrics.classification_report(y_test, y_pred, labels=[1, 0]))
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
```

```python
y_pred = xgb.predict(X_test)
score = accuracy_score(y_test, y_pred)
score
```

```python
print(metrics.classification_report(y_test, y_pred, labels=[1, 0]))
```
